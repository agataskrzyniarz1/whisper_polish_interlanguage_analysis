```python
import json
import re
```

```python
# Chargement du lexique polonais sous forme d'ensemble de mots
with open("lexique_polonais.txt", encoding="utf-8") as f:
    lexique_polonais = set(word.strip().lower() for word in f if word.strip())

# Chargement des transcriptions à partir du fichier JSON
with open("transcriptions_ordonnées.json", encoding="utf-8") as f:
    data = json.load(f)
```

```python
# Fonction pour nettoyer et découper le texte en mots
def tokenize(text):
    text = text.lower()
    text = re.sub(r"[^\wąćęłńóśźż]+", " ", text)  # Retirer la ponctuation (mais garder les caractères polonais)
    return text.strip().split()

# Ensemble pour stocker les mots qui ne sont pas dans le lexique polonais
mots_pas_polonais = set()

# Parcourir les transcriptions automatiques en polonais uniquement
for item in data:
    transcription = item.get("transcription automatique")
    if transcription:  # Ignorer celles qui sont marquées comme non polonaises
        mots = tokenize(transcription)
        for mot in mots:
            if mot not in lexique_polonais:
                mots_pas_polonais.add(mot)

# Affichage des mots non trouvés dans le lexique
print("Mots de la transcription automatique qui ne sont pas en polonais :")
for mot in sorted(mots_pas_polonais):
    print(mot)

# Sauvegarde des mots non reconnus dans un fichier texte
with open("automatique_mots_pas_polonais.txt", "w", encoding="utf-8") as f:
    for mot in sorted(mots_pas_polonais):
        f.write(mot + "\n")
```
